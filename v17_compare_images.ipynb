```python
import math
import matplotlib.pyplot as plt
import pickle
import random
```

```python
num_of_images = 50
```

```python
# CREATE A LIST OF RANDOM SEEDS TO BE USED ON ORIGINAL AND DEBIASED SCRIPTS
rnd_list = random.sample(range(999,1999), num_of_images) # [52,49,47]
with open('rnd_list.pkl', 'wb') as f:
    pickle.dump(rnd_list, f)
print(f'Number of Unique seeds: {len(set(rnd_list))}')
```

```python
# AFTER BOTH SCRIPTS RAN AND CREATED EACH LIST OF IMAGES, RUN BELOW TO PLOT THE COMPARISON
plt.rcParams['figure.figsize'] = (40*0.8,num_of_images*5*0.8)
plt.rcParams['axes.titlesize'] = 'large'
plt.rcParams['axes.labelsize'] = 'large'
f, axarr = plt.subplots(int(num_of_images/2),4)

with open('original_imgs.pkl', 'rb') as f:
    original_imgs = pickle.load(f)

with open('debiased_imgs.pkl', 'rb') as f:
    debiased_imgs = pickle.load(f)

for i in range(len(original_imgs)):
    row = math.floor(i/2)
    if (i % 2) == 0: col = 0
    else: col = 2
    axarr[row,col].set_title(original_imgs[i]['rnd'])
    axarr[row,col].imshow(original_imgs[i]['img'])
    axarr[row,col].set_xlabel('Original Image')
    axarr[row,col+1].set_title(debiased_imgs[i]['rnd'])
    axarr[row,col+1].imshow(debiased_imgs[i]['img'])
    axarr[row,col+1].set_xlabel('Debiased Image')
```
